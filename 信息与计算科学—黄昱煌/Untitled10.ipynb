```python
import cv2
import numpy as np
import matplotlib.pyplot as plt #
image_path = 'pic.jpg'
original_image = cv2.imread(image_path)#
image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)# BGR  RGB
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')#
denoised_image = cv2.medianBlur(original_image, 5)#
denoised_image_rgb = cv2.cvtColor(denoised_image, cv2.COLOR_BGR2RGB)#bgrrgb
plt.subplot(1, 2, 2)
plt.imshow(denoised_image_rgb)
plt.title('makeImage')
plt.axis('off')
cv2.imwrite('filter_turing.png', denoised_image_rgb)#()
plt.show()#
```

```python
def linear_stretching(image):
    min_val = np.min(image)
    max_val = np.max(image)
    stretched_image = ((image - min_val) / (max_val - min_val) * 255).astype(np.uint8)
    return stretched_image#
plt.subplot(1,1, 1)
plt.imshow(stretched_image, cmap='gray')
plt.title('Stretched Image')
plt.axis('off')
plt.savefig('Stretched Image')#
plt.show()
```

```python
tx, ty = 50, 30 
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])#

translated_image = cv2.warpAffine(original_image, translation_matrix, (original_image.shape[1], original_image.shape[0]))#

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title('Translated Image')#()
plt.axis('off')
plt.savefig('Translated Image')
plt.show()   #
```

```python
flipped_image_horizontal = cv2.flip(original_image, 1)  # 
flipped_image_vertical = cv2.flip(original_image, 0)    # 
flipped_image_both = cv2.flip(original_image, -1)       # 
plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(flipped_image_horizontal, cv2.COLOR_BGR2RGB))
plt.title('Flipped Horizontal')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(flipped_image_vertical, cv2.COLOR_BGR2RGB))
plt.title('Flipped Vertical')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(flipped_image_both, cv2.COLOR_BGR2RGB))
plt.title('Flipped Both')
plt.axis('off')
plt.savefig('Overturn')#
plt.show()
```

```python
target_size = (original_image.shape[1] //2, original_image.shape[0] //1)# 

resized_image = cv2.resize(original_image, target_size, interpolation=cv2.INTER_LINEAR)# 

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.axis('off')
plt.savefig('Resized Image')#
plt.show()# 


```
